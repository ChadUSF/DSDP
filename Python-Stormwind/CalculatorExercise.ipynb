## Calculator
Create a program that performs some basic math.

Your Calculator program should be able to:

Menu
1. Add two numbers
2. Substract two numbers
3. Divide two numbers
4. Multiple two numbers
5. Calculate 8% tax on a value
6. Calculate monthly payments, given a total amount and a number of years 
7. Quit

It should run in an infinite loop until the user enters 7. Use the following skeleton code.


```python
while True:
    #print out the menu
    print(
    '''
    Calculator v1.0 Menu
    1. Add two numbers
    2. Substract two numbers
    3. Divide two numbers
    4. Multiple two numbers
    5. Calculate 8% tax on a value
    6. Calculate monthly payments, given a total amount and a number of years 
    7. Quit
    
    ''')
    
    #print user input
    userChoice = int(input('Please choose an option from 1 - 6: '))
    
    #if user chooses 1, add 2 numbers
    if userChoice == 1:
        print('ADD TWO NUMBERS')
        number_1 = float(input('First number: '))
        number_2 = float(input('Second number: '))
        result = number_1 + number_2
        print('Result: ', result)

    #else if user chooses 2, substract 2 numbers
    elif userChoice == 2:
        print('SUBSTRACT TWO NUMBERS')
        number_1 = float(input('First number: '))
        number_2 = float(input('Second number: '))
        result = number_1 - number_2
        print('Result: ', result)

    # else if user chooses 3, divide 2 numbers
    elif userChoice == 3:
        print('DIVIDING TWO NUMBERS')
        number_1 = float(input('First number: '))
        number_2 = float(input('Second number: '))
        result = number_1 / number_2
        print('Result: ', result)

    # else if user chooses 4, multiply 2 numbers
    elif userChoice == 4:
        print('MULTIPLY TWO NUMBERS')
        number_1 = float(input('First number: '))
        number_2 = float(input('Second number: '))
        result = number_1 * number_2
        print('Result: ', result)
    
    # else if user chooses 5, calculate 8% tax
    elif userChoice == 5:
        print('CALCULATING 8% TAX ON A VALUE')
        amount = float(input('Enter amount: '))
        result = amount * 0.08
        print('Result: ', result)
    
    # else if user chooses 6, calculate monthly payment
    elif userChoice == 6:
        print('CALCULATING MONTHLY PAYMENTS')
        amount = float(input('Enter loan value: '))
        number_of_years = float(input('Enter number of years: '))
        payment = amount / (number_of_years * 12)
        print('Result: ', payment)
       
    # else if user choose 7, program quit.
    elif userChoice == 7:
        print("Thank you, goodbye!")
        break
```
